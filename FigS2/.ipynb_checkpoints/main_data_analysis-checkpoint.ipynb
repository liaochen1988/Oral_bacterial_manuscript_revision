```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from statsmodels.stats.multitest import multipletests
from scipy.stats import mannwhitneyu
```

```python
df_bek = pd.read_csv("metadata.csv", index_col=0)

# SGD/TGD starts 10 days before bone marrow transplant day
df_bek['Timepoint'] = [int(day.replace('Day','').strip())-10 for day in df_bek.Timepoint]
df_bek['Antibiotic'] = [abx.split('--')[0] for abx in df_bek.Antibiotic]
df_bek['AntibioticWhen'] = [int(abxw.split('From day 0 to day ')[1].split(',')[0])-10 for abxw in df_bek.AntibioticWhen]

# since the data is sparse, bin the data first
my_bins = np.array([-10, 0, 20, 40, 1000])
df_bek['BinIndex'] = np.digitize(df_bek.Timepoint, bins=my_bins)
df_bek['BinAbx'] = ['%d_%s'%(b,a) for b,a in zip(df_bek['BinIndex'], df_bek['Antibiotic'])]
print(Counter(df_bek['BinIndex']))
binlabels = ['Day [%d,%d)'%(my_bins[k],my_bins[k+1]) for k,b in enumerate(my_bins) if k!=len(my_bins)-1]

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6,4), sharey=True)
PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'k'},
    'medianprops':{'color':'r'},
    'whiskerprops':{'color':'k'}
}
_ = sns.boxplot(x='BinIndex', y='OralFraction', hue='Antibiotic', ax=ax, data=df_bek, **PROPS, showfliers=False)
_ = sns.stripplot(x='BinIndex', y='OralFraction', hue='Antibiotic', ax=ax, data=df_bek, size=6, color=".3", linewidth=0, dodge=True, palette='Set2', jitter=0.2)
_ = ax.set_xticklabels(binlabels)
```

```python
for bi in [1,2,3,4]:
    res = mannwhitneyu(
        x=df_bek[(df_bek.BinIndex==bi) & (df_bek.Antibiotic=='SGD(oral)')].OralFraction, 
        y=df_bek[(df_bek.BinIndex==bi) & (df_bek.Antibiotic=='TGD(oral)')].OralFraction,
        alternative='less'
    )
    print(bi, res)
```

```python
multipletests([0.023, 1.82e-10, 3.2e-4, 5.2e-4], method='fdr_bh')
```

```python
for binindex in [1,2,3,4]:
    df2 = df_bek.loc[df_bek.BinIndex==binindex, ['Antibiotic','OralFraction']]
    
    # Variances
    variances = df2.groupby('Antibiotic').var(ddof=1)
    # Mean variance
    mean_var = variances.mean().values[0]
    # Pooled standard deviation
    s_pooled = np.sqrt(mean_var)
    # Difference of the means
    diff_mean = abs(df2.groupby('Antibiotic').mean().diff().values[-1])
    # Cohen's d
    cohens_d = diff_mean / s_pooled
    
    print(binindex, "Cohen's d = %.3f"%(cohens_d))
```
