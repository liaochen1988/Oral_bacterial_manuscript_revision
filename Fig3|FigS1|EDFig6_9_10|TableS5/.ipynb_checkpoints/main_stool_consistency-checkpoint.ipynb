```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib.patches as mpatches
import seaborn as sns
from copy import deepcopy
%config InlineBackend.figure_format = 'retina'
import numpy as np
from statsmodels.stats.multitest import multipletests
from collections import Counter
import sys
import warnings
import pingouin as pg
from scipy.stats import ttest_ind, wilcoxon, ttest_1samp, ttest_rel, kruskal, wilcoxon, mannwhitneyu
from scipy.stats import spearmanr
if not sys.warnoptions:
    warnings.simplefilter("ignore")
from statannot import add_stat_annotation
```

```python
#read sample info
df_sample = pd.read_csv('tblASVsamples.csv', index_col=0)

# read count data
df_count = pd.read_csv('tblcounts_asv_melt.csv')
df_count = pd.pivot_table(df_count, index='SampleID', columns='ASV', values='Count', aggfunc=np.sum).fillna(0)
sample_ids_w_suff_reads = list(df_count.sum(axis=1)[df_count.sum(axis=1)>=1000].index)
df_count = df_count.loc[list(set(df_sample.index).intersection(set(sample_ids_w_suff_reads)))]
df_relab =  df_count.div(df_count.sum(axis=1), axis=0)

# get oral bacterial fraction
df_blast_100 = pd.read_csv("blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_100.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']#
df_oral_total = df_relab[list(set(df_blast_100.query_accver).intersection(df_relab.columns))].sum(axis=1).to_frame()
df_oral_total.columns = ['OralFrac']
df_sform = pd.merge(df_oral_total, df_sample, how='inner', left_index=True, right_index=True)
df_sform = df_sform[df_sform.Consistency.notnull()]

ndata_formed = len(df_sform[df_sform.Consistency=='formed'])
ndata_semi = len(df_sform[df_sform.Consistency=='semi-formed'])
ndata_liquid = len(df_sform[df_sform.Consistency=='liquid'])
print('formed stool # = %d, semi-formed stool # = %d, liquid stool # = %d'%(ndata_formed, ndata_semi, ndata_liquid))
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4.5,4.5))
plt.rcParams.update({'font.size': 16})
_ = sns.stripplot(x='Consistency', y='OralFrac', data=df_sform, ax=ax, order=['formed','semi-formed','liquid'], palette='Set2', jitter=0.1, s=2)
_ = sns.boxplot(x="Consistency", y="OralFrac", data=df_sform, saturation=1, showfliers=False, width=0.5, palette='Set2', boxprops={'zorder': 30, 'facecolor': 'none'}, ax=ax, order=['formed','semi-formed','liquid'])
                
# # use df2_sform for statistical test
# _ = add_stat_annotation(ax, data=df_sform, x='Consistency', y='log10_OralFrac', order=['formed','semi-formed','liquid'],
#                         box_pairs=[("formed", "semi-formed"), ("formed", "liquid"), ("semi-formed", "liquid")],
#                         comparisons_correction=None,
#                         test='Kruskal', text_format='star', loc='outside', verbose=1)

plt.tight_layout()
plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("stool_consistency_oral_correlation.svg")
```

```python
res = mannwhitneyu(df_sform[df_sform.Consistency=='formed'].OralFrac, df_sform[df_sform.Consistency=='semi-formed'].OralFrac)
print("formed vs semi-formed", res)
res = mannwhitneyu(df_sform[df_sform.Consistency=='semi-formed'].OralFrac, df_sform[df_sform.Consistency=='liquid'].OralFrac)
print("semi-formed vs liquid", res)
res = mannwhitneyu(df_sform[df_sform.Consistency=='formed'].OralFrac, df_sform[df_sform.Consistency=='liquid'].OralFrac)
print("formed vs liquid", res)
```

```python
for pair in [('formed','semi-formed'),('semi-formed','liquid'),('formed','liquid')]:
    df2 = df_sform.loc[(df_sform.Consistency==pair[0])|(df_sform.Consistency==pair[1]), ['Consistency','OralFrac']]
    
    # Variances
    variances = df2.groupby('Consistency').var(ddof=1)
    # Mean variance
    mean_var = variances.mean().values[0]
    # Pooled standard deviation
    s_pooled = np.sqrt(mean_var)
    # Difference of the means
    diff_mean = abs(df2.groupby('Consistency').mean().diff().values[-1])
    # Cohen's d
    cohens_d = diff_mean / s_pooled
    
    print(pair, "Cohen's d = %.3f"%(cohens_d))
```
