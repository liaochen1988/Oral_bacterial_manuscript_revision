```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
from statsmodels.stats.multitest import multipletests
import warnings
warnings.filterwarnings('ignore')
```

# Compute absolute abundance

```python
# top genera with most >30% samples
(pd.read_csv("Liao_alloHCT_relabun_absabun_genera.csv", index_col=0).iloc[:,4:-2]>0.3).astype(int).T.sum(axis=1).sort_values().tail(10)
```

```python
df = pd.read_csv("Liao_alloHCT_relabun_absabun_genera.csv", index_col=0)
selected_genera = ['Blautia','Erysipelatoclostridium','Bifidobacterium','[Clostridium] innocuum group',
                   'Akkermansia','Bacteroides','Enterococcus','Klebsiella',
                   'Escherichia-Shigella','Lactobacillus','Streptococcus']
df = df[selected_genera + ['OralFraction','qPCR16S']]
df = df[df.qPCR16S >= 1e3]
df['log10_qPCR16S'] = np.log10(df.qPCR16S)
for genus in selected_genera + ['OralFraction']:
    df['abs_'+genus] = df[genus] * df['qPCR16S']
    df['log10_rel_'+genus] = np.log10(df[genus])
    df['log10_abs_'+genus] = np.log10(df['abs_'+genus])
df.head()
```

# Plot relative abundance vs absolute abundance

```python
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(15,6), sharex=True, sharey=True)

for k, genus in enumerate(selected_genera[:-1]):
    k1 = int(k/5)
    k2 = k%5
    
    df2 = df[df[genus] > 0]
    df2 = df2[df2.log10_qPCR16S >= 3]
    _ = sns.regplot(data=df2, x='log10_rel_'+genus, y='log10_qPCR16S', ax=ax[k1,k2], scatter_kws={"color": "black"}, line_kws={"color": "red"}, scatter=False)
    _ = sns.scatterplot(data=df2, x='log10_rel_'+genus, y='log10_qPCR16S', ax=ax[k1,k2], color='lightgray', edgecolor='lightgray', legend=False, s=6)

    # linear regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(df2['log10_rel_'+genus], df2['log10_qPCR16S'], alternative='less')
    print(genus)
    print('n=', len(df2))
    print('slope = %2.2f, r = %2.2f, P = %2.2e, std_err = %2.2f'%(slope, r_value, p_value, std_err))
    
    _ = ax[k1,k2].set_title('n=%d, slope=%2.2f, P=%2.2e'%(len(df2), slope, p_value))
    _ = ax[k1,k2].set_xlabel('Relative abundance of ' + genus)
    _ = ax[k1,k2].set_ylabel('Total bacterial load')

plt.tight_layout()
plt.rcParams['svg.fonttype'] = 'none'
fig.savefig("rel_abundance_vs_total_bacterial_load.svg", dpi=600)
```
