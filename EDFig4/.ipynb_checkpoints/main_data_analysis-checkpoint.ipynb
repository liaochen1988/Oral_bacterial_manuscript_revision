```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from statsmodels.stats.multitest import multipletests
from scipy.stats import mannwhitneyu
from copy import deepcopy
import warnings
warnings.filterwarnings("ignore")
```

# Load metadata, relative abundance, and oral ASVs

```python
df_meta = pd.read_csv("metadata.csv", index_col=0)
df_meta.head()
```

```python
df_relabun = pd.read_csv("relative_abundance_wide_format.csv", index_col=0)
df_relabun.head()
```

```python
df_blast_HMPv35 = pd.read_csv("blast_HMPv35_p100.txt", sep="\t", comment="#", header=None)
df_blast_HMPv35.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
df_blast_HMPv35.head()
```

```python
df_blast_HMPv35oral = pd.read_csv("blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_HMPv35oral.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']
df_blast_HMPv35oral.head()
```

# Extended Data Fig. 4a

```python
df_cutoff = pd.read_csv("HMP_oral_asvs_varied_cutoffs.csv", index_col=0)

df_oral_frac = None
for col in df_cutoff.columns:
    relabun_cutoff = float(col.split(', ')[0])
    prevalence_cutoff = float(col.split(', ')[1])
    all_oral_asvs = list(df_cutoff[df_cutoff[col]==1].index)
    hits = list(set(df_blast_HMPv35[df_blast_HMPv35.subject_accver.isin(all_oral_asvs)].query_accver).intersection(set(df_relabun.columns)))
    oral_frac_in_stool = df_relabun.loc[list(set(df_meta[df_meta.BodySite=='stool'].index).intersection(set(df_relabun.index))), hits].sum(axis=1).to_frame().rename({0:col}, axis=1)
    if df_oral_frac is None:
        df_oral_frac = deepcopy(oral_frac_in_stool)
    else:
        df_oral_frac = pd.merge(df_oral_frac, oral_frac_in_stool, left_index=True, right_index=True, how='inner')
df_oral_frac = pd.merge(df_oral_frac, df_meta[['Disease']], left_index=True, right_index=True, how='left').sort_values('Disease')
df_oral_frac.columns = [(',').join(['%s%%' % str(float(x)*100) for x in col.split(', ')]) if col != 'Disease' else 'Disease' for col in df_oral_frac.columns]
df_oral_frac = df_oral_frac.set_index('Disease', drop=True)

fig, ax = plt.subplots(figsize=(12,3))
_ = sns.heatmap(
    df_oral_frac.T,
    vmin=0,
    vmax=0.3,
    cmap='Spectral_r',
    yticklabels=True,
    ax=ax
)
```

# Extended Data Fig. 4a-c

```python
df_blast_HMPv35oral = pd.read_csv("blast_HMPv35oral_p100.txt", sep="\t", comment="#", header=None)
df_blast_HMPv35oral.columns = ['query_accver', 'subject_accver', 'perc_identity', 'alignment_length', 'mismatches', 'gap_opens', 'qstart', 'qend', 'sstart', 'send', 'evalue', 'bitscore']

res = []
all_subjects = list(set(df_meta.SubjectID))
for sub in all_subjects:
    
    # find paired sample IDs
    curr_df = df_meta[df_meta.SubjectID==sub]
    gut_sample_id = list(curr_df[curr_df.BodySite=='stool'].index)
    assert len(gut_sample_id) == 1
    gut_sample_id = gut_sample_id[0]
    oral_sample_id = list(curr_df[curr_df.BodySite=='saliva'].index)
    assert len(oral_sample_id)==1
    oral_sample_id = oral_sample_id[0]
    
    if gut_sample_id not in list(df_relabun.index) or oral_sample_id not in list(df_relabun.index):
        continue
    else:
        disease = list(curr_df.Disease)[0]
    
    # oral sequences (sequences present in the oral cavity of patients)
    df_oral_seqs = df_relabun.loc[oral_sample_id]
    oral_seqs = list(set(df_oral_seqs[df_oral_seqs>0].index))
    
    # fecal sequences (sequences present in the fecal cavity of patients)
    df_fecal_seqs = df_relabun.loc[gut_sample_id]
    fecal_seqs = list(set(df_fecal_seqs[df_fecal_seqs>0].index))
    
    # oral bacterial fraction in fecal sample (reference-based approach)
    HMPv35oral_hit = list(set(df_blast_HMPv35oral.query_accver).intersection(set(df_relabun.columns)))
    oral_frac_in_feces_reference = df_relabun.loc[gut_sample_id, HMPv35oral_hit].sum()
    
    # oral bacterial fraction in fecal sample (based on overlap)
    seq_overlap = list(set(oral_seqs).intersection(fecal_seqs))
    oral_frac_in_feces_overlap = df_relabun.loc[gut_sample_id, seq_overlap].sum()
    
    # proportion of the number of inferred oral ASVs that are present in paired oral cavity
    fecal_asvs=list(df_relabun.loc[gut_sample_id][df_relabun.loc[gut_sample_id]>0].index)
    fecal_asvs_inferred_as_oral = list(set(fecal_asvs).intersection(set(df_blast_HMPv35oral.query_accver)))
    assert np.abs(oral_frac_in_feces_reference-df_relabun.loc[gut_sample_id,fecal_asvs_inferred_as_oral].sum())<1e-6
    
    if len(fecal_asvs_inferred_as_oral) == 0:
        res.append([gut_sample_id, 0, 0, 0, np.NaN, np.NaN, np.NaN, disease])
    else:
        fecal_asvs_inferred_as_oral_and_found_in_oral_cavity = list(set(fecal_asvs_inferred_as_oral).intersection(oral_seqs))
        proportion_of_oral_asv_number = len(fecal_asvs_inferred_as_oral_and_found_in_oral_cavity)/len(fecal_asvs_inferred_as_oral)
        proportion_of_oral_asv_fraction = df_relabun.loc[gut_sample_id, fecal_asvs_inferred_as_oral_and_found_in_oral_cavity].sum()/df_relabun.loc[gut_sample_id, fecal_asvs_inferred_as_oral].sum()
        res.append([gut_sample_id, len(fecal_asvs_inferred_as_oral), len(fecal_asvs_inferred_as_oral_and_found_in_oral_cavity),
                    oral_frac_in_feces_reference, oral_frac_in_feces_overlap, proportion_of_oral_asv_number, proportion_of_oral_asv_fraction, disease])

df_res = pd.DataFrame(res, columns=['SampleID','NumOralASV','NumOralAVSInOralCavity','OralFracRef','OralFracOverlap','PercOralASVInOralCavity','FracInOralSample','Disease']).set_index('SampleID')
df_res.head()
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, figsize=(8,3))

df2plot = df_res[df_res.OralFracRef > 0]
PROPS = {
    'boxprops':{'facecolor':'none', 'edgecolor':'black'},
    'medianprops':{'color':'black'},
    'whiskerprops':{'color':'black'},
    'capprops':{'color':'black'}
}

# oral bacterial fraction
_ = sns.boxplot(x="Disease", y="OralFracRef", data=df2plot, ax=ax[0], palette='Set2', order=["HC", "CD","UC"], showfliers=False, **PROPS)
_ = sns.stripplot(x="Disease", y="OralFracRef", data=df2plot, ax=ax[0], s=8.5, edgecolor='k', linewidth=0.5, palette='Set2', order=["HC", "CD","UC"], jitter=True)
_ = ax[0].set_yscale('log')

# percentage of fraction present in the oral cavity
_ = sns.boxplot(x="Disease", y="PercOralASVInOralCavity", data=df2plot, ax=ax[1], palette='Set2', order=["HC", "CD","UC"], showfliers=False, **PROPS)
_ = sns.stripplot(x="Disease", y="PercOralASVInOralCavity", data=df2plot, ax=ax[1], s=8.5, edgecolor='k', linewidth=0.5, palette='Set2', order=["HC", "CD","UC"], jitter=True)

# percentage of fraction present in the oral cavity
_ = sns.boxplot(x="Disease", y="FracInOralSample", data=df2plot, ax=ax[2], palette='Set2', order=["HC", "CD","UC"], showfliers=False, **PROPS)
_ = sns.stripplot(x="Disease", y="FracInOralSample", data=df2plot, ax=ax[2], s=8.5, edgecolor='k', linewidth=0.5, palette='Set2', order=["HC", "CD","UC"], jitter=True)

plt.tight_layout()
```

```python
for disease in ['CD', 'UC']:
    df2 = df_res.loc[(df_res.Disease=='HC') | (df_res.Disease==disease), ['Disease','OralFracRef']]
    
    # Variances
    variances = df2.groupby('Disease').var(ddof=1)
    # Mean variance
    mean_var = variances.mean().values[0]
    # Pooled standard deviation
    s_pooled = np.sqrt(mean_var)
    # Difference of the means
    diff_mean = abs(df2.groupby('Disease').mean().diff().values[-1])
    # Cohen's d
    cohens_d = diff_mean / s_pooled
    
    print(disease, "Cohen's d = %.3f"%(cohens_d))
```
